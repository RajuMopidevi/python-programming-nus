<b>To write a program:</b>
1. Break the problem down into a series of steps (pseudocode)
2. Convert each step into Python code

That seems simple but can be difficult to do because the program may not work.

<b>What can go wrong?</b>
1. Steps may not be executed in the correct order. If the steps are correct but the order is wrong, then your code will not work.
2. A step may be left out. You may have it solved mentally but you may leave out a step.
3. A step may be mistranslated into code. Your pseudocode is correct but one or more of them was improperly translated.

<b>Consider this problem:</b><br>
Get two numbers from the user and add them to produce a result.<br>
If the result is even, display "Result is EVEN!", else display "Result is ODD!".

<b>Consider the steps:</b>
1. Get a number from the user.
2. Get another number from the user.
3. Add the two numbers entered to produce a result.
4. Print result for verification purpose.
4. Work out the remainder when the result is divided by 2.
5. If the remainder is 0, display "Result is EVEN!", else display "Result is ODD!".

# Writing Pseudocode
Produce the steps as pseudocode 

```python
# Prompt user for a number
# Prompt user for another number
# Add the two numbers entered by user to produce a result
# Print result for verification purpose
# Work out the remainder when the result is divided by 2
# If remainder is 0 then
#    print "Result is EVEN!"
# else
#    print "Result is ODD!"
```

# Writing Python Code
Convert pseudocode to Python code 

```python
# Prompt user for a number
number1 = input("Enter first number: ")

# Prompt user for another number
number2 = input("Enter second number: ")

# Add the two numbers entered by user to produce a result
result = number1 + number2

# Print result for verification purpose
print("Result is", result)

# Work out the remainder when the result is divided by 2
remainder = result % 2

# If remainder is 0 then
if remainder == 0:
    
#    print "Result is EVEN!"
    print("Result is EVEN!")
    
# else
else:
    
#    print "Result is ODD!"    
    print("Result is ODD!")
```



The two numbers, 3 and 4, entered are correct values but why does Python report it as error?<br>
Notice the result is `34` and not `7`.<br>
This is a case where a step is mistranslated in the code.<br>
The problem is due to the input() function which prompts the user for input and returns what the user typed.<br>
The problem lies in what input() returns.

# A Working Code

```python
# Prompt user for a number
number1 = int(input("Enter first number: "))

# Prompt user for another number
number2 = int(input("Enter second number: "))

# Add the two numbers entered by user to produce a result
result = number1 + number2

# Print result for verification purpose
print("Result is", result)

# Work out the remainder when the result is divided by 2
remainder = result % 2

# If remainder is 0 then
if remainder == 0:
    
#    print "Result is EVEN!"
    print("Result is EVEN!")
    
# else
else:
    
#    print "Result is ODD!"    
    print("Result is ODD!")
```

**Observations**<br>
The int() function converts what the user typed into an integer value.<br>
The result is now `7`.<br>
Since the remainder from dividing `result` by `2` is not 0, the `else` part of the `if` statement is executed.
